## Mengukur Jarak Data Iris

```python
import pandas as pd
from scipy.spatial.distance import euclidean, cityblock

# membaca dataset
df = pd.read_csv('iris.csv')

# mengambil hanya data numerik
data_numerik = df[['sepal_length','sepal_width','petal_length','petal_width']]

# mengambil dua data pertama
data1 = data_numerik.iloc[0]
data2 = data_numerik.iloc[1]

# menghitung jarak Euclidean
jarak_euclidean = euclidean(data1, data2)

# menghitung jarak Manhattan
jarak_manhattan = cityblock(data1, data2)

print("Jarak Euclidean:", jarak_euclidean)
print("Jarak Manhattan:", jarak_manhattan)
```

## Data Campuran

```python
df_mix = df.copy()

# Ordinal (ukuran bunga)
df_mix['Size'] = pd.cut(df_mix['sepal_length'],
                        bins=[0,5,6,10],
                        labels=['small','Medium','Large'])

# Binary (lebar petal besar atau tidak)
df_mix['WidePetal'] = (df_mix['petal_width'] > 1).astype(int)

df_mix.head()
```

## Hasil Transform data

Data Campuran:
1. Data Numeric
- sepal_length
- sepal_width
- petal_length
- petal_width

2. data Ordinal:
- Small
- Medium
- Large

3. Biner:
- WidePetal

Data yang berisi 0 = Petal Kecil, 1 = Petal Lebar

4. Kategorikal:
- Iris-setosa
- Iris-versicolor
- Iris-virginica

## Menghitung Jarak Data (Euclidean Distance)

```python
from scipy.spatial.distance import pdist, squareform

numeric_data = df[['sepal_length','sepal_width','petal_length','petal_width']]

distance_matrix = pd.DataFrame(
    squareform(pdist(numeric_data, metric='euclidean'))
)

distance_matrix.head()
```

Interpretasi:

- Nilai 0 menunjukkan jarak data dengan dirinya sendiri.

- Semakin kecil nilai jarak, semakin mirip karakteristik bunga tersebut.

- Semakin besar nilai jarak, semakin berbeda karakteristik bunga.

Contoh:

- Jarak data 0 dan 1 = 0.54
Artinya kedua bunga memiliki ukuran yang cukup mirip tetapi tidak identik.

- Jarak data 1 dan 2 = 0.30
Artinya bunga tersebut lebih mirip dibanding pasangan lainnya.

## Pengukuran Jarak Data Campuran

```python
import pandas as pd
!pip install gower
import gower

# Create mixed_data directly from df_mix
mixed_data = df_mix[['sepal_length','petal_width','species','Size','WidePetal']]

# Define which columns are categorical for Gower
# The order must match the columns in 'mixed_data'
# sepal_length (False), petal_width (False), species (True), Size (True), WidePetal (True)
cat_features_gower = [False, False, True, True, True]

gower_dist = gower.gower_matrix(mixed_data, cat_features=cat_features_gower)

pd.DataFrame(gower_dist).head()
```

1. Matriks menunjukkan jarak antar data

Setiap angka pada tabel menunjukkan tingkat perbedaan antara dua data bunga iris.

Contoh interpretasi:

- Baris 0 kolom 1 = 0.08
Artinya data bunga ke-0 dan ke-1 memiliki perbedaan yang sangat kecil, sehingga karakteristiknya hampir mirip.

- Baris 0 kolom 3 = 0.15
Artinya data bunga ke-0 dan ke-3 memiliki perbedaan yang lebih besar dibanding pasangan sebelumnya.

2. Nilai diagonal selalu 0
- setiap data dibandingkan dengan dirinya sendiri
- sehingga jaraknya = 0


3. Rentang nilai Gower Distance

0  → sangat mirip

1  → sangat berbeda

Pada dataset iris biasanya nilai jarak cukup kecil, karena banyak bunga memiliki ukuran yang hampir sama.

4. Jarak dihitung dari data campuran

Perhitungan jarak pada tabel ini mempertimbangkan beberapa tipe data sekaligus, yaitu:

- Numeric :(sepal_length, petal_width)
- Kategorikal :(species)
- Ordinal :(Size (Small, Medium, Large))
- Binary :(WidePetal (0 atau 1))

Karena data terdiri dari berbagai tipe, digunakan metode Gower Distance yang memang dirancang untuk data campuran.
