```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Definisi Data
x = np.array([2, 4, 5, 3, 3, 4, 5])
y = np.array([2, 3, 5, 4, 3, 5, 6])

# Perhitungan Analitik (Manual)
n = len(x)
m_analitik = (n * np.sum(x * y) - np.sum(x) * np.sum(y)) / (n * np.sum(x**2) - np.sum(x)**2)
c_analitik = (np.sum(y) - m_analitik * np.sum(x)) / n

# Perhitungan dengan Sklearn
model = LinearRegression().fit(x.reshape(-1, 1), y)

print(f"Analitik: y = {m_analitik:.4f}x + {c_analitik:.4f}")
print(f"Sklearn:  y = {model.coef_[0]:.4f}x + {model.intercept_:.4f}")
```
